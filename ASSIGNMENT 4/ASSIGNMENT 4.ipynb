```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
print("Dataset Shape:", df.shape)
```

```python
print("\nInfo:")
print(df.info())
```

```python
df.describe(include='all')
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100
missing_data = pd.DataFrame({'Missing Values': missing, 'Percentage (%)': missing_percent})
print(missing_data)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
num_features = ['Age', 'Fare']
df[num_features].hist(figsize=(12,5), bins=30, color='skyblue')
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(12,5))
for i, col in enumerate(num_features, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
cat_features = ['Pclass', 'Sex', 'Embarked', 'Survived']
plt.figure(figsize=(14,6))
for i, col in enumerate(cat_features, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=df, x=col, palette='Set2', hue=None)
    plt.title(f'Count Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x='Sex', y='Survived', palette='Set1')
plt.title("Survival Rate by Sex")
plt.show()
```

```python
sns.barplot(data=df, x='Pclass', y='Survived', palette='Set3')
plt.title("Survival Rate by Pclass")
plt.show()
```

```python
sns.violinplot(data=df, x='Survived', y='Age', palette='muted')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.kdeplot(data=df[df['Survived']==1]['Fare'], label='Survived', shade=True)
sns.kdeplot(data=df[df['Survived']==0]['Fare'], label='Not Survived', shade=True)
plt.legend()
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
cross_tab = pd.crosstab(df['Sex'], df['Pclass'], values=df['Survived'], aggfunc='mean')
plt.figure(figsize=(6,4))
sns.heatmap(cross_tab, annot=True, fmt=".2f", cmap='Blues')
plt.title("Survival Rate by Sex and Pclass")
plt.show()
```

```python
df_corr = df.copy()
df_corr['Sex'] = df_corr['Sex'].map({'male': 0, 'female': 1})
```

```python
corr_matrix = df_corr.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Fare'] < (Q1 - 1.5 * IQR)) | (df['Fare'] > (Q3 + 1.5 * IQR))]
print("Number of Fare outliers:", outliers.shape[0])
```

```python
# Outlier detection using IQR for SibSp
Q1 = df['SibSp'].quantile(0.25)
Q3 = df['SibSp'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['SibSp'] < (Q1 - 1.5 * IQR)) | (df['SibSp'] > (Q3 + 1.5 * IQR))]
print("Number of SibSp outliers:", outliers.shape[0])
```
